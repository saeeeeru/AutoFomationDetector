## トラッキングデータの解析

[The DEBS 2013 Grand Challeng](https://www2.informatik.uni-erlangen.de/publication/download/DEBS2013b.pdf)で提案されているデータセットを使用。  
[情報とサッカー](https://tetsuyatyamamoto.blogspot.com/)を参考にしながら解析を進める。  
このデータの解析の目的は以下の2つである。  
1. 守備時の選手の矢印（大きさと方向を持つ）を描画できるようにする
2. 攻撃のパターンでなく、守備時のパターンを発見する

#### 必要なライブラリのインポート

```python
import os, random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import japanize_matplotlib
```

#### データの読み込み

```python
infile = os.path.join('.', 'full-game')

# n = sum(1 for line in open(infile)) - 1
n = 49576080

p = 0.01
n_rows = int(n/4)
df = pd.read_csv(infile, 
#                  header=None)
                 header=None, 
                 nrows=n_rows)
#                  skiprows=lambda i: i>0 and random.random() > p)

df.columns = ['sid', 'ts', 'x', 'y', 'z', '|v|', '|a|', 'vx', 'vy', 'vz', 'ax', 'ay', 'az']

df.iloc[:,1:].describe()
```

#### データの理解（センサーID）

```python
print('センサー数 :{}'.format(len(df['sid'].unique())))
sid_list = np.sort(df['sid'].unique()).tolist()
print('センサーID :{}'.format(sid_list))

# レフリー、ボール、各プレイヤーのセンサーID
left_r = 105
right_r = 106

ball_list = [4, 8, 10, 12]  # 12は後半のみ
left_a_list = [13, 47, 49, 19, 53, 23, 57, 59]
right_a_list = [14, 16, 88, 52, 54, 24, 58, 28]

left_b_list = [61, 63, 65, 67, 69, 71, 73, 75]
right_b_list = [62, 64, 66, 68, 38, 40, 74, 44]
```

#### データの理解（タイムスタンプ）
タイムスタンプは17桁で、秒の小数部が12桁ある。

```python
t_array = np.sort(df['ts'].unique())
print('タイムスタンプ数: {}'.format(len(t_array)))

# 1 time-tickでどれだけのセンサーのデータを収集できているか確認
print('when t = {} , data of {} sensors is collected'.format(t_array[0], len(df[df.ts == t_array[0]].sid.unique())))

# 試合開始時からのデータとする
# from datetime import datetime

df['ts'] = [float(str(t)[:5]+'.'+str(t)[5:]) for t in df.ts.values.tolist()]
st = 10753.295594424116
df = df[df.ts > st]
df.ts -= st

print('final time-stamp: {}'.format(np.sort(df['ts'].unique())[-1]))
# print('N = {} -> N_tmp = {}'.format(len(df), len(df_tmp)))

summary = df.groupby(['sid']).count()['ts']

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
fig.suptitle('各プレイヤーのデータ点数')

left_list = left_a_list + left_b_list
summary[left_list].plot(kind='bar', rot=0, color='darkgrey', title='左足', ax=axes[0])
axes[0].set_ylabel('cnt')

right_list = right_a_list + right_b_list
summary[right_list].plot(kind='bar', rot=0, color='darkgrey', title='右足', ax=axes[1])
axes[1].set_ylabel('cnt')

# 各センサーの初めのtime-stampと、deltaの平均を表示
print('started arrival of time: {}'.format(st))
print('sensor: initial time-stamp, end time-stamp, mean of timedelta')
t_list = df[df.sid == ball_list[0]].ts.values.tolist()
dt_list = [t_list[i]-t_list[i-1] for i in range(1,len(t_list)-1)]
print('Ball: {}, {}, {}'.format(t_list[0], t_list[-1], np.mean(dt_list)))
for i, leg_list in enumerate([[left_a_list,right_a_list],[left_b_list, right_b_list]]):
    for j, sid_list in enumerate(leg_list):
        for k, sid in enumerate(sid_list):
            t_list = df[(df.sid == sid)].ts.values.tolist()
            dt_list = [t_list[i]-t_list[i-1] for i in range(1,len(t_list)-1)]
            print('Number {}s {} of Team {}: {}, {}, {}'.format(k+1,['left', 'right'][j], ['A', 'B'][i], t_list[0], t_list[-1], np.mean(dt_list)))
```

```python
from matplotlib.patches import Arc, Rectangle, ConnectionPatch

# ボールと選手の平均位置情報を確認
cmap = plt.get_cmap('tab10')

def draw_pitch():
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))

    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)

    ax.plot([33965, -33960], [0, 50], color='black')
    ax.plot([-33968, -33960], [-22560, 50], color='black')
    ax.plot([-33968, -33968], [-29880, -22560], color='black', linewidth=10)
    ax.plot([-33968, -33939], [-29880, -52489], color='black')
    ax.plot([-33939, 33941], [-52489, -52477], color='black')
    ax.plot([33941, 33941], [-52477, -29898], color='black')
    ax.plot([33941, 33941], [-29898, -22578], color='black', linewidth=10)
    ax.plot([33941, 33965], [-22578, 0], color='black')

    ax.plot([0, 0], [0, -52477], color='black')
    centreCircle = plt.Circle((0, -52477/2), 7000, color="black",fill=False)
    ax.add_patch(centreCircle)

    return fig, ax
    
fig, ax = draw_pitch()

# 平均的なポジションの描画

# df_tmp[(df_tmp.sid == ball_list[0])].plot(x='x', y='y', ax=ax, color='red', legend=False)

# ax.scatter(x=df[(df.sid == ball_list[0])].x.mean(), y=df[(df.sid == ball_list[0])].y.mean(),
ax.scatter(x=df[(df.sid == ball_list[0])].y.mean(), y=-df[(df.sid == ball_list[0])].x.mean(),
           color='red',
           s=250)

for i, left_list in enumerate([left_a_list, left_b_list]):
    for j, left in enumerate(left_list):
#         df_tmp[(df_tmp.sid == left)].plot(x='x', y='y', ax=ax, color=cmap(i), legend=False)
        x, y = df[(df.sid == left)][['x', 'y']].mean().tolist()
#         ax.scatter(x=x, y=y,
        ax.scatter(x=y, y=-x,
                   color=cmap(i),
#                    marker='^' if i==0 else 'v',
                   marker='>' if i==0 else '<',
                   s=500)
        
#         ax.annotate('{}'.format(j+1), (x, y), size=12, ha='center', va='center', color='white')
        ax.annotate('{}'.format(j+1), (y, -x), size=12, ha='center', va='center', color='white')

# ax.legend(['ball']+['a {}'.format(i) for i in range(len(left_a_list))] + ['b {}'.format(i) for i in range(len(left_b_list))])
```

#### ボール保持者と保持チーム、その距離、ボールの速度の分析

```python
import time
from matplotlib.ticker import ScalarFormatter

dt = 0.25
T = 200
threshold_ball = 0.5e+7
threshold_distance = 2e+3

class FixedOrderFormatter(ScalarFormatter):
    def __init__(self, order_of_mag=0, useOffset=True, useMathText=True):
        self._order_of_mag = order_of_mag
        ScalarFormatter.__init__(self, useOffset=useOffset, 
                                 useMathText=useMathText)
    def _set_orderOfMagnitude(self, range):
        self.orderOfMagnitude = self._order_of_mag

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 12))
axes[0].set_xlim([0, T*dt]); axes[0].set_ylabel('ボール保持者とそのチーム')
axes[1].set_xlim([0, T*dt]); axes[1].set_ylabel('ボールと保持者の距離')
axes[2].set_xlim([0, T*dt]); axes[2].set_ylabel('ボールの速度')
axes[2].yaxis.set_major_formatter(FixedOrderFormatter(7 ,useMathText=True))

t_list, d_list, v_list = [], [], []

st = time.time()
for t in range(0,T):
    img_list= []
    st_tmp, ed_tmp = t*dt, (t+1)*dt
    
    ball_features = df[(st_tmp <= df.ts)&(df.ts < ed_tmp)&(df.sid == ball_list[0])][['x', 'y','vx','vy','|v|']].mean().values
   
    d_array = np.array([[np.linalg.norm(df[(st_tmp <= df.ts)&(df.ts < ed_tmp)&((df.sid == left)|(df.sid == right))][['x', 'y']].mean().values-ball_features[:2]) for (left, right) in zip(left_list, right_list)] for (left_list, right_list) in zip([left_a_list, left_b_list], [right_a_list, right_b_list])])
    tid, pid = np.unravel_index(d_array.argmin(), d_array.shape)
    min_d = d_array.min()
    
    # detect pass moving
    if ball_features[-1] > threshold_ball and t != 0:
        tid = pre_tid
        
        pid = pid if min_d < threshold_distance else pre_pid
        
    t_list.append(st_tmp); d_list.append(min_d); v_list.append(ball_features[-1])

    axes[0].annotate('{}'.format(pid+1), (st_tmp, tid), ha='center', va='center', color=cmap(tid))    
    
    pre_tid, pre_pid = tid, pid

axes[1].plot(t_list, d_list)
axes[2].plot(t_list, v_list)

print('elapsed time: {}[sec]'.format(time.time()-st))
```

```python
%matplotlib nbagg
import matplotlib.animation as animation
import time

dt = 0.25
T = int(600/dt)
threshold_pass = 0.5e+7
threshold_distance = 2e+3

fig, ax = draw_pitch()
art_list = []

st = time.time()

for t in range(0,T):
    img_list= []
    st_tmp, ed_tmp = t*dt, (t+1)*dt
    
    ball_features = df[(st_tmp <= df.ts)&(df.ts < ed_tmp)&(df.sid == ball_list[0])][['x', 'y','vx','vy','|v|']].mean().values
    
    d_array = np.array([[np.linalg.norm(df[(st_tmp <= df.ts)&(df.ts < ed_tmp)&((df.sid == left)|(df.sid == right))][['x', 'y']].mean().values-ball_features[:2]) for (left, right) in zip(left_list, right_list)] for (left_list, right_list) in zip([left_a_list, left_b_list], [right_a_list, right_b_list])])
    tid, pid = np.unravel_index(d_array.argmin(), d_array.shape)
    min_d = d_array.min()
    
    # here, we must detect type of ball possession　based on velocity of ball, and distance between player and ball
    if ball_features[-1] > threshold_pass and t != 0:
        # for direct pass or ball controll
        pid = pid if min_d < threshold_distance and tid == pre_tid else pre_pid
        tid = pre_tid
        flag = True if df[(ed_tmp <= df.ts)&(df.ts < (t+2)*dt)&(df.sid == ball_list[0])][['|v|']].mean().values <= threshold_pass else False
    else:
        flag = True
    
    img_list.append(plt.text(0.5, 1.01, '[1st half] {} ~ {} [sec] = Num {}{} of Team {}'.format(st_tmp, ed_tmp, pid+1, '' if flag else '\' pass', tid), horizontalalignment='center', verticalalignment='bottom', transform=ax.transAxes))
    img_list.append(ax.scatter(x=ball_features[1], y=-ball_features[0],
               color='red',
               s=250))
#     img_list.append(ax.quiver(ball_features[1], -ball_features[0], ball_features[3], -ball_features[2], color='red', animated=True))
    pre_tid, pre_pid = tid, pid

    for i, (left_list, right_list) in enumerate(zip([left_a_list, left_b_list], [right_a_list, right_b_list])):
        for j, (left, right) in enumerate(zip(left_list, right_list)):
            x, y, vx, vy = df[(st_tmp <= df.ts)&(df.ts < ed_tmp)&((df.sid == left)|(df.sid == right))][['x', 'y', 'vx', 'vy']].mean().values
            img_list.append(ax.scatter(x=y, y=-x,
                                       color=cmap(i),
                                       marker='>' if i==0 else '<',
                                       s=500,
                                       edgecolors='black' if flag and (i==tid and j==pid) else 'face'))
            
            if i != tid:
                img_list.append(ax.quiver(y, -x, vy, -vx, color=cmap(i), animated=True))
            img_list.append(ax.annotate('{}'.format(j+1), (y, -x), size=12, ha='center', va='center', color='white'))
    
    art_list.append(img_list)

anim = animation.ArtistAnimation(fig, art_list)
anim.save("output_possession.gif", writer="pillow")
print('elapsed time: {}[sec]'.format(time.time()-st))
```
